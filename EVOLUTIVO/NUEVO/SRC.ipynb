```python
import pandas as pd

# Nombre del archivo en la misma carpeta
file_name = "EVOLUCION-ALQ-MALAGA-CIUDAD.xlsx"

# Cargar el archivo Excel
df = pd.read_excel(file_name, sheet_name=None, dtype=str)  # Cargar todas las hojas

# Definir los valores a añadir
ciudad = "Malaga"
area = "Centro"  # Modifica según sea necesario

# Función para agregar columnas a cada hoja
def add_columns(sheet):
    sheet["Ciudad"] = ciudad
    sheet["Área"] = area
    return sheet

# Aplicar la función a todas las hojas
updated_sheets = {name: add_columns(sheet) for name, sheet in df.items()}

# Guardar el archivo modificado con un nuevo nombre
output_file = "EVOLUCION-ALQ-MALAGA-CIUDAD.xlsx"
with pd.ExcelWriter(output_file, engine="xlsxwriter") as writer:
    for sheet_name, updated_df in updated_sheets.items():
        updated_df.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
# Nombre del archivo en la misma carpeta
file_name = "EVOLUCION-ALQ-MALAGA-CIUDAD.xlsx"

# Cargar el archivo Excel con todas sus hojas
df_sheets = pd.read_excel(file_name, sheet_name=None, dtype=str)

# Definir los valores a añadir
ciudad = "Malaga"
area = "Ciudad"  # Modifica según sea necesario

# Función para procesar cada hoja
def process_sheet(sheet):
    sheet = sheet.fillna("")  # Reemplazar NaN con cadena vacía

    # Separar la primera columna por espacios y expandir en nuevas columnas
    split_cols = sheet.iloc[:, 0].str.split(expand=True)
    
    # Combinar con el resto del DataFrame (si hay más columnas)
    sheet = pd.concat([split_cols, sheet.iloc[:, 1:]], axis=1)

    # Agregar las nuevas columnas "Ciudad" y "Área"
    sheet["Ciudad"] = ciudad
    sheet["Area"] = area
    
    return sheet

# Aplicar la función a todas las hojas
processed_sheets = {name: process_sheet(sheet) for name, sheet in df_sheets.items()}

# Guardar el archivo modificado
output_file = "EVOLUCION-ALQ-MALAGA-CIUDAD.xlsx"
with pd.ExcelWriter(output_file, engine="xlsxwriter") as writer:
    for sheet_name, processed_df in processed_sheets.items():
        processed_df.to_excel(writer, sheet_name=sheet_name, index=False)
```
