```python
import pandas as pd
import numpy as np

# Cargar el archivo Excel
file_path = "COM-MAD-CIUDAD1.xlsx"  # Asegúrate de colocar la ruta correcta
df = pd.read_excel(file_path, sheet_name="Sheet_1", dtype=str)

# Reemplazar NaN por cadenas vacías para evitar errores
df = df.fillna("")

# Función para mover datos a la izquierda fila por fila
def shift_row_left(row):
    values = [x for x in row if x.strip() != ""]  # Filtrar valores vacíos
    empty_count = len(row) - len(values)  # Contar cuántos valores vacíos se eliminaron
    return values + [""] * empty_count  # Rellenar con espacios vacíos al final

# Aplicar la función a cada fila del DataFrame
df_left_aligned = pd.DataFrame(df.apply(shift_row_left, axis=1).to_list(), columns=df.columns)

# Eliminar filas que solo tienen información en la primera columna
df_filtered = df_left_aligned[df_left_aligned.iloc[:, 1:].apply(lambda row: any(row.str.strip() != ""), axis=1)]

# Guardar el archivo con los datos alineados y filtrados
df_filtered.to_excel("COM-MAD-CIUDAD1.xlsx", index=False)
```
