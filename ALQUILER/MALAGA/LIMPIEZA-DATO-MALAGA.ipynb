```python
import pandas as pd

# Cargar el archivo Excel
file_path = "ALQ-MALG-BENAL2.xlsx"  # Reemplaza con el nombre correcto
df = pd.read_excel(file_path, sheet_name="Sheet_1", dtype=str)

# Reemplazar NaN por cadenas vacías
df = df.fillna("")

# Función para eliminar celdas que contienen "https"
def remove_https(row):
    return ["" if "https" in str(x) else x for x in row]

# Aplicar la función a cada fila
df_cleaned = pd.DataFrame(df.apply(remove_https, axis=1).to_list(), columns=df.columns)

# Función para mover datos a la izquierda fila por fila
def shift_row_left(row):
    values = [x for x in row if x.strip() != ""]  # Filtrar valores vacíos
    empty_count = len(row) - len(values)  # Contar cuántos valores vacíos se eliminaron
    return values + [""] * empty_count  # Rellenar con espacios vacíos al final

# Aplicar la función para alinear los datos a la izquierda
df_final = pd.DataFrame(df_cleaned.apply(shift_row_left, axis=1).to_list(), columns=df.columns)

# Guardar el archivo modificado
df_final.to_excel("ALQ-MALG-BENAL2.xlsx", index=False)
```
